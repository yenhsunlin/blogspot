```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import seaborn as sns
sns.set()
#%config InlineBackend.figure_format = 'retina'
```

```python
# Generating scattered points inside unit square uniformly
X,Y=np.random.uniform(-1,1,size=(2,1000))
```

```python
# Plot
fig = plt.figure()
ax = fig.add_subplot(111)

cir1 = Circle(xy = (0.0, 0.0),linewidth=5 ,radius=1,color="r",fill=False)
ax.add_patch(cir1)

plt.scatter(X,Y,alpha=0.5)
plt.axis('scaled')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.axes().set_aspect('equal')
#plt.savefig("unit_circle",dpi=300,bbox_inches = 'tight')
```

```python
def unitcircle(N):
    """
    Generating points on the unit circle.
    N: How many points we want to generate?
    """
    
    # Step 1
    X,Y = np.random.uniform(-1,1,size=(2,N))
    # Step 2: Normalization constant
    r = np.sqrt(X**2+Y**2)
    
    return X/r,Y/r
```

```python
# Generating 100 points distributed on a unit circle uniformly
dx,dy=unitcircle(100)
```

```python
# Plot
fig = plt.figure()
ax = fig.add_subplot(111)

cir1 = Circle(xy = (0.0, 0.0),linewidth=2 ,radius=1,color="r",fill=False)
ax.add_patch(cir1)

plt.scatter(dx,dy,alpha=0.6)

plt.xlabel("$x$")
plt.ylabel("$y$")
plt.axes().set_aspect('equal')
#plt.savefig("unit_circle_pts.png",dpi=300,bbox_inches = 'tight')
```

```python
def nsphere(N,dim=2):
    """
    Generating points on the unit sphere in arbitrary dimensions.
    N: How many points we want to generate?
    dim: Dimension of the sphere.
    """
    
    # Step 1
    X = np.random.uniform(-1,1,size=(dim,N))
    # Step 2: Normalization constant
    r = np.linalg.norm(X,axis=0)
    
    # The following returns:
    # 1st row: all N points' positions on x-axis
    # 2nd row: all N points' positions on y-axis
    # ...
    # d-th row: all N points' positions on d-axis
    # It can be projected to traditional (x,y,z,...,d) coordinate by transpose T
    return X/r
```

```python
# Test nsphere can generate identical result as unitcircle
x,y = nsphere(2000)
```

```python
plt.scatter(x,y,alpha=0.05)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.axes().set_aspect('equal')
```
